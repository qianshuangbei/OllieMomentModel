# Yolo Segment

Test Segment mode


```python
import cv2
import os
from ultralytics import YOLO
from IPython.display import Image, display

os.makedirs('predict', exist_ok=True)

# Load a model
model = YOLO("yolo11n-seg.pt")  # load an official model
# model = YOLO("path/to/best.pt")  # load a custom model

# Predict with the model
results = model("image/sv2_000017.jpg")  # predict on an image

# Plot results and save
for r in results:
    # Plot segmentation masks with colored overlay
    im_array = r.plot()  # plot a BGR numpy array of predictions
    
    # Get original filename without path
    orig_filename = os.path.basename("image/sv2_000017.jpg")
    save_path = os.path.join('predict', f'seg_{orig_filename}')
    
    # Save the visualization
    cv2.imwrite(save_path, im_array)
    display(Image(filename=save_path))
```
