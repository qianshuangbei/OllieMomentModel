# Yolo Detect

This notebook reads videos from the 'video' directory and samples frames at 5 FPS

```python
import cv2
import os
from glob import glob
from tqdm import tqdm
```

```python
# Set parameters
VIDEO_DIR = 'video'
OUTPUT_DIR = 'image'
TARGET_FPS = 10

# Create output directory if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
def extract_frames(video_path):
    # Get video name without extension
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    
    # Read video
    cap = cv2.VideoCapture(video_path)
    
    # Get video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Calculate frame interval to achieve target FPS
    interval = int(fps / TARGET_FPS)
    
    frame_idx = 0
    saved_count = 0
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
            
        if frame_idx % interval == 0:
            # Save frame
            output_path = os.path.join(OUTPUT_DIR, f"{video_name}_{saved_count:06d}.jpg")
            cv2.imwrite(output_path, frame)
            saved_count += 1
            
        frame_idx += 1
    
    cap.release()
    return saved_count
```

```python
# Get list of video files
video_files = glob(os.path.join(VIDEO_DIR, '*.mp4'))
print(f"Found {len(video_files)} video files")

# Process each video
for video_path in tqdm(video_files):
    print(f"\nProcessing {video_path}...")
    frames_saved = extract_frames(video_path)
    print(f"Saved {frames_saved} frames")
```

```python
import cv2
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load a model
model = YOLO("yolo11n-cls.pt")  # load an official model
# model = YOLO("path/to/best.pt")  # load a custom model

# Predict with the model
results = model("image/sv2_000017.jpg")  # predict on an image

for r in results:
    # Read the original image
    img = cv2.imread("image/sv2_000017.jpg")
    
    # Get prediction info
    probs = r.probs  # cls prob (list)
    pred_class = probs.top1  # predicted class id
    conf = float(probs.top1conf)  # confidence score
    class_name = model.names[pred_class]  # class name
    
    # Add text to image
    text = f'{class_name}: {conf:.2f}'
    cv2.putText(img, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Get original filename without path
    orig_filename = os.path.basename("image/sv2_000017.jpg")
    save_path = os.path.join('predict', f'detect_{orig_filename}')
    
    # Save the visualization
    cv2.imwrite(save_path, img)
    
    # Display result (for Jupyter notebook)
    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    
    # Print classification results
    print(f'Predicted class: {class_name}')
    print(f'Confidence: {conf:.2f}')
```
